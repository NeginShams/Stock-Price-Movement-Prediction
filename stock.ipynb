```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import json

def scrape_sahamyab_stocktwits():
    # Setup Chrome options
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Run in background
    options.add_argument('--disable-gpu')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    
    # driver = webdriver.Chrome(options=options)
    # driver.get("https://www.sahamyab.com/stocktwits")
    
    # Wait for initial load
    try:
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.twits-card"))
        )
    except TimeoutException:
        print("Page didn't load properly")
        driver.quit()
        return
    
    # Initialize variables
    last_height = driver.execute_script("return document.body.scrollHeight")
    collected_tweets = []
    seen_ids = set()
    scroll_attempts = 0
    max_attempts = 5
    
    while scroll_attempts < max_attempts:
        # Scroll to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Adjust based on your internet speed
        
        # Get current scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        # Find all tweet cards
        tweet_cards = driver.find_elements(By.CSS_SELECTOR, "div.twits-card")
        
        # Process new tweets
        new_items_found = False
        for card in tweet_cards:
            try:
                tweet_id = card.get_attribute("id")
                if tweet_id and tweet_id not in seen_ids:
                    # Extract tweet content
                    content = card.find_element(By.CSS_SELECTOR, "p.twitsContents").text
                    
                    # Extract additional information
                    username = card.find_element(By.CSS_SELECTOR, "span.twits-username").text
                    timestamp = card.find_element(By.CSS_SELECTOR, "span.twits-time").text
                    
                    # Extract hashtags
                    hashtags = [tag.text for tag in card.find_elements(By.CSS_SELECTOR, "a.tweet-url")]
                    
                    # Extract symbols if available
                    symbols = [symbol.text for symbol in card.find_elements(By.CSS_SELECTOR, "a.symbol")]
                    
                    tweet_data = {
                        "id": tweet_id,
                        "username": username,
                        "content": content,
                        "timestamp": timestamp,
                        "hashtags": hashtags,
                        "symbols": symbols
                    }
                    
                    collected_tweets.append(tweet_data)
                    seen_ids.add(tweet_id)
                    new_items_found = True
                    
            except NoSuchElementException:
                continue
        
        # Check if we've reached the end
        if new_height == last_height and not new_items_found:
            scroll_attempts += 1
        else:
            scroll_attempts = 0
        
        last_height = new_height
    
    # Save results to JSON file
    with open('sahamyab_tweets.json', 'w', encoding='utf-8') as f:
        json.dump(collected_tweets, f, ensure_ascii=False, indent=2)
    
    print(f"Scraping complete. Collected {len(collected_tweets)} tweets.")
    # driver.quit()

if __name__ == "__main__":
    scrape_sahamyab_stocktwits()
```
